```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("saude_do_sono_estilo_vida.csv")
df
```

```python
# exercício 1
df.rename(columns={'ID': 'Identificador',
                   'Pressão sanguíneaaaa' : 'pressão sanguínea' ,
                   'Ocupação' : 'Profissão',
                   'Categoria BMI' : 'Categoria IMC'}, inplace = True)
df
```

```python
# Exercício 2
media_profissao = df.groupby('Profissão')['Duração do sono'].mean()
media_profissao
```

```python
mediana_profissao = df.groupby('Profissão')['Duração do sono'].median()
mediana_profissao
```

```python
moda_profissao = df.groupby('Profissão')['Duração do sono'].apply(lambda x: x.mode().iloc[0])
moda_profissao
```

```python
# Exercício 3
engenheiro = df[df['Profissão'] == 'Eng. de Software']
engenheiro.shape
```

```python
obesos = (engenheiro['Categoria IMC'] == 'Obesidade').sum()
obesos
```

```python
total_eng = len(engenheiro)
total_eng
```

```python
porcentagem_obesos = (obesos / total_eng) * 100
porcentagem_obesos
```

```python
#Exercício 4
advogado_representante = df[df['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])]
advogado_representante.shape
```

```python
media_advogado_representante = advogado_representante['Duração do sono'].mean()
media_advogado_representante
```

```python
media_profissoes = df['Duração do sono'].mean()
media_profissoes
```

```python
if media_advogado_representante < media_profissoes:
    print("Advogar ou ser representante de vendas faz você dormir menos.")
elif media_advogado_representante > media_profissoes:
    print("Advogar ou ser representante de vendas não faz você dormir menos.")
else:
    print("Não há diferença significativa na média de horas de sono entre advogar ou ser representante de vendas e todas as profissões.")
```

```python
#Exercício 5
media_sono_enfermeiro = df[df['Profissão'].isin(['Enfermeiro(a)'])]['Duração do sono'].mean()
media_sono_enfermeiro
```

```python
media_sono_medico = df[df['Profissão'].isin(['Médico(a)'])]['Duração do sono'].mean()
media_sono_medico
```

```python
if media_sono_enfermeiro > media_sono_medico:
  print('Enfermeiros têm mais horas de sono')
else:
  print('Médicos têm mais horas de sono')
```

```python
#Exercício 8
df[['Pressão Sistólica', 'Pressão Diastólica']] = df['pressão sanguínea'].str.split('/', expand=True)
```

```python
df['Pressão Sistólica'] = pd.to_numeric(df['Pressão Sistólica'])
df['Pressão Diastólica'] = pd.to_numeric(df['Pressão Diastólica'])
```

```python
media_mulher = df[df['Gênero'] == 'Mulher'][['Pressão Sistólica', 'Pressão Diastólica']].mean().mean()
```

```python
media_homem = df[df['Gênero'] == 'Homem'][['Pressão Sistólica', 'Pressão Diastólica']].mean().mean()
```

```python
if media_homem > media_mulher:
    print("Os homens têm uma pressão sanguínea maior.")
elif media_homem < media_mulher:
    print("As mulheres têm uma pressão sanguínea maior.")
else:
    print("Homens e mulheres têm a mesma pressão sanguínea.")
```

```python
#Exercício 9
horas_sono_moda = df['Qualidade do sono'].mode()
horas_sono_moda
```

```python
if 8 in horas_sono_moda.values:
    print("É predominante dormir 8 horas por dia.")
else:
    print("Não é predominante dormir 8 horas por dia.")
```

```python
#Exercício 10
media_passos_acima_70 = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()
media_passos_acima_70
```

```python
media_passos_70_menos = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()
media_passos_70_menos
```

```python
if media_passos_acima_70 > media_passos_70_menos:
    print("Pessoas com frequências cardíacas acima de 70 dão mais passos.")
elif media_passos_acima_70 < media_passos_70_menos:
    print("Pessoas com frequências cardíacas igual ou menor a 70 dão mais passos.")
else:
    print("Não há diferença na média de passos.")
```
